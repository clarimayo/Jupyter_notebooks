# Exercise 1

Plot one chart using matplotlib (OOP-style) using the following data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

open - where it opened at on that data
high - the highest point it hit that day
low - the lowest point it hit that day
close - where the price closed the day at
adj close - closing price adjusted for stock splits and dividends

```python
df = pd.read_csv('GOOG.csv')
df
```

```python
df.info()
```

```python
df.groupby('Date')[["Open", "High", "Low", "Close", "Volume"]].sum().sort_values(by="", ascending=False)[1:10].plot.barh(stacked=True)
```

```python
dates = tuple([i for i in df["Date"].value_counts().keys()])[:10]
```

```python
values = df[["Open", "High", "Low", "Close"]][:10]
```

```python
x = np.linspace(0,2,100)

fig, ax = plt.subplots()

ax.plot(dates,values)
ax.set_xlabel("Date")
ax.set_ylabel("Records")
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set_title("Stock Market Records")
plt.legend(["Open", "High", "Low", "Close"])
```

# Exercise 2

Perform some initial exploratory analysis to the following dataframe. Then plot two or more charts related to the data.

```python
df = pd.read_csv('SHSAT-Registrations-and-Testers.csv')
df
```

```python
df.head()
```

```python
#checking to see if there are any missing values
df.info()
```

```python
ax = df.groupby("School name")[["Grade level","Number of students who registered for the SHSAT", "Number of students who took the SHSAT"]].sum().sort_values(by="Grade level", ascending=False)[:10].plot.barh(stacked=True, figsize=(12,8))
ax.set_ylabel("Schools")
ax.set_xlabel("Number of students and their grade level")
ax.set_title("Records of students based on school")
plt.show()
```

```python
#seeing what type of relationship the columns have
sns.pairplot(df)
plt.show()
```

```python
#it seems there is a positive correlation between "Number of students who registered for the SHSAT" 
#and "Number of students who took the SHSAT"
plt.figure(figsize=(5,5))
sns.jointplot(x=df["Number of students who registered for the SHSAT"].values, y=df["Number of students who took the SHSAT"].values,height=7)
plt.ylabel("Students who took the SHSAT")
plt.xlabel("Students who registered for the SHSAT")
plt.show()
```

```python
#confirmed the positive relation. Will do linear regression between these two values
df.corr() 
```

```python
X = df["Number of students who registered for the SHSAT"]
Y = df["Number of students who took the SHSAT"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train.values.reshape(-1,1), y_train.values.reshape(-1,1))
```

```python
score = reg.score(X_test.values.reshape(-1,1), y_test.values.reshape(-1,1))
```

```python
x_line = np.arange(100).reshape(-1,1)

ax=sns.regplot(x="Number of students who registered for the SHSAT", y="Number of students who took the SHSAT", data=df, fit_reg=True)
plt.plot(x_line, reg.predict(x_line))
m,b = np.polyfit(X,Y,1)
print(m)
print(b)
plt.show()
```

```python
f"The equation of this line is y={reg.intercept_} + {reg.coef_}x"
```
